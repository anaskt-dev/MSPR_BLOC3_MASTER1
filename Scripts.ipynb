```python

# --- Chemins ---
base = r"D:\MSPR_CODE\MSPR_BLOC3_MASTER1"

f1 = base + r"\data_empo\base-cc-emploi-pop-active-2022.csv"
f2 = base + r"\DATA\Second_tour_rhone_sep.csv"
f3 = base + r"\DATA\50_MTS-52_CHO_sheet1.csv"
f4 = base + r"\DATA\donnee-dep-data.gouv-2025-geographie2025-produit-le2026-01-22.csv"

out = base + r"\DATA"
```

```python
df = pd.read_csv(f1, sep=";", encoding="utf-8")


codgeo_num = pd.to_numeric(df["CODGEO"], errors="coerce")

df_rhone = df[(codgeo_num >= 1024) & (codgeo_num <= 1536)]

print(df_rhone.head())
print("Nombre de lignes Rhône :", len(df_rhone))

df_rhone.to_csv("base_rhone_filtre.csv",sep = ";",encoding="utf-8" ,index=False)
```

```python

# --- 4. Big fichier géographie (filtrer Rhône = 69) ---
geo = pd.read_csv(f4, sep=";", encoding="utf-8")
geo_rhone = geo[geo["Code_departement"].astype(str) == "69"]
geo_rhone = geo_rhone.reset_index(drop=True)
```
