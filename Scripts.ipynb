```python

# --- Chemins ---
base = r"D:\MSPR_CODE\MSPR_BLOC3_MASTER1"

f1 = base + r"\DATA\Premier_tour_rhone_sep.csv"
f2 = base + r"\DATA\Second_tour_rhone_sep.csv"
f3 = base + r"\DATA\50_MTS-52_CHO_sheet1.csv"
f4 = base + r"\DATA\donnee-dep-data.gouv-2025-geographie2025-produit-le2026-01-22.csv"

out = base + r"\data_normalisation\dataset_rhone.csv"
```

```python

# --- 1. Premier tour (déjà Rhône) ---
df1 = pd.read_csv(f1, sep=";", encoding="utf-8")


# --- 2. Second tour (déjà Rhône) ---
df2 = pd.read_csv(f2, sep=";", encoding="utf-8")
```

```python
# Pour déboguer tes colonnes CHO juste après la lecture:
test = pd.read_csv(f3, sep=";", encoding="utf-8", header=None, dtype=str, engine="python")
print(test.head(6))
```

```python
# --- 3) CHÔMAGE --- Fix correct ---
cho = pd.read_csv(f3, sep=";", encoding="utf-8", header=None, dtype=str)

# utiliser la vraie ligne d'en-tête (ligne 1)
cho.columns = cho.iloc[1]

# supprimer les lignes 0 et 1
cho = cho.iloc[2:].reset_index(drop=True)

# filtrer Rhône = 69
cho_rhone = cho[cho["Code département"].astype(str).str.strip() == "69"].reset_index(drop=True)

print(cho_rhone)
```

```python

# --- 4. Big fichier géographie (filtrer Rhône = 69) ---
geo = pd.read_csv(f4, sep=";", encoding="utf-8")
geo_rhone = geo[geo["Code_departement"].astype(str) == "69"]
geo_rhone = geo_rhone.reset_index(drop=True)
```

```python

# --- 5. Concaténer proprement ---
df_concat = pd.concat([df1, df2, cho_rhone, geo_rhone], ignore_index=True)
```

```python

# --- 6. Sauvegarde ---
os.makedirs(os.path.dirname(out), exist_ok=True)
df_concat.to_csv(out, index=False, sep=";", encoding="utf-8")

print("✔ Dataset Rhône créé :", out)
print("Lignes finales :", len(df_concat))
```
